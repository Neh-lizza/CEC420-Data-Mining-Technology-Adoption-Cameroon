## Purpose

This notebook generates the student survey dataset used in the main project notebook.

The survey was designed using Google Forms and targets Cameroonian university students. The questions cover AI usage habits and over-reliance behavior across four regions: Buea, Bamenda, Yaounde, and Douala.

The dataset is structured to match the exact column names and answer options of the live Google Form. Response distributions are based on realistic assumptions about Cameroonian university student behavior.

**Output:** `data/survey_responses_synthetic.csv` — 657 rows x 15 columns

**Run this notebook first before running the main project notebook.**

---
## Step 1 — Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

np.random.seed(42)
os.makedirs('data', exist_ok=True)

n = 657  # total responses
print(f'Generating {n} survey responses...')
```

---
## Step 2 — Section 1: Demographics

Region, field of study, academic level, and gender. Buea and Bamenda are weighted higher because the University of Buea is the primary technology institution in Cameroon.

```python
# Timestamps spread across Jan–Mar 2025 to simulate real collection
timestamps = pd.date_range('2025-01-15', periods=n, freq='47min')

# Q1 — Region
region = np.random.choice(
    ['Buea', 'Bamenda', 'Yaounde', 'Douala', 'Other'],
    n, p=[0.30, 0.25, 0.20, 0.15, 0.10]
)

# Q2 — Field of study
field = np.random.choice(
    ['Engineering', 'Sciences', 'Social Sciences', 'Law', 'Health', 'Business', 'Other'],
    n, p=[0.30, 0.18, 0.12, 0.10, 0.10, 0.12, 0.08]
)

# Q3 — Academic level
level = np.random.choice(
    ['100L', '200L', '300L', '400L', 'Masters', 'PhD'],
    n, p=[0.15, 0.25, 0.28, 0.20, 0.08, 0.04]
)

# Q4 — Gender
gender = np.random.choice(['Male', 'Female'], n, p=[0.55, 0.45])

print('Demographics generated.')
print(f'  Regions   : {dict(zip(*np.unique(region, return_counts=True)))}')
print(f'  Gender    : {dict(zip(*np.unique(gender, return_counts=True)))}')
```

---
## Step 3 — Section 2: AI Usage Habits

How many tools students use, how many hours per day, and what tasks they use AI for. Usage of 3–4 hours per day is the most common range based on current student behavior observations.

```python
# Q5 — Number of AI tools used
tools = np.random.choice(
    ['0', '1', '2', '3', '4 or more'],
    n, p=[0.05, 0.15, 0.30, 0.28, 0.22]
)

# Q6 — Daily hours using AI
hours = np.random.choice(
    ['Less than 1hr', '1-2hrs', '3-4hrs', '5hrs or more'],
    n, p=[0.12, 0.30, 0.35, 0.23]
)

# Q7 — Academic tasks AI is used for
tasks = np.random.choice(
    ['Assignments', 'Research', 'Summarising notes', 'Coding', 'Exam prep', 'Other'],
    n, p=[0.25, 0.20, 0.18, 0.20, 0.12, 0.05]
)

print('AI usage habits generated.')
print(f'  Daily hours: {dict(zip(*np.unique(hours, return_counts=True)))}')
```

---
## Step 4 — Section 3: Over-Reliance Indicators

These are the key variables used to compute the risk score. Scales are 1–5 throughout. Low verification and high anxiety indicate over-reliance.

```python
# Q8 — How often verify AI output (1=Never, 5=Always)
# Weighted toward 3-4: most students sometimes verify
verify = np.random.choice([1, 2, 3, 4, 5], n, p=[0.08, 0.15, 0.27, 0.30, 0.20])

# Q9 — Confidence without AI (1=Not at all, 5=Very confident)
# Weighted low: many students feel lost without AI
confidence = np.random.choice([1, 2, 3, 4, 5], n, p=[0.22, 0.28, 0.25, 0.15, 0.10])

# Q10 — Submitted AI content as own work
submitted = np.random.choice(
    ['Never', 'Once or twice', 'Sometimes', 'Often'],
    n, p=[0.30, 0.30, 0.25, 0.15]
)

# Q11 — Anxious when AI unavailable (1=Never, 5=Always)
anxious = np.random.choice([1, 2, 3, 4, 5], n, p=[0.10, 0.18, 0.28, 0.27, 0.17])

# Q12 — Trust AI without checking source (1=Never, 5=Always)
trust = np.random.choice([1, 2, 3, 4, 5], n, p=[0.08, 0.15, 0.30, 0.28, 0.19])

print('Over-reliance indicators generated.')
```

---
## Step 5 — Section 4: Awareness

AI literacy and whether the student's university has provided guidance on responsible AI use.

```python
# Q13 — Understands AI limitations (1=Not at all, 5=Very well)
literacy = np.random.choice([1, 2, 3, 4, 5], n, p=[0.10, 0.20, 0.30, 0.25, 0.15])

# Q14 — University gave guidance on responsible AI use
# 50% say no — reflects current Cameroon university reality
guidance = np.random.choice(['yes', 'no', 'not sure'], n, p=[0.25, 0.50, 0.25])

print('Awareness section generated.')
print(f'  University guidance: {dict(zip(*np.unique(guidance, return_counts=True)))}')
```

---
## Step 6 — Build DataFrame

Column names match the exact Google Form question text so the dataset looks identical to a real Google Forms CSV export.

```python
df = pd.DataFrame({
    'Timestamp'                                                                      : timestamps,
    'What region do you attend university'                                           : region,
    'What is your field of study?'                                                   : field,
    'What academic level are you at?'                                                : level,
    'What is your gender?'                                                           : gender,
    'How many AI tools do you use regularly? (ChatGPT, Gemini, Claude, Copilot, etc.)': tools,
    'How many hours per day do you spend using AI tools?'                            : hours,
    'For which academic tasks do you use AI? (select all that apply)'                : tasks,
    'How often do you verify what an AI tool tells you before using it?'             : verify,
    'How confident are you completing academic tasks without AI?'                    : confidence,
    'Have you submitted AI-generated content as your own work without major edits?'  : submitted,
    'Do you feel anxious or stuck when AI tools are unavailable?'                    : anxious,
    'How much do you trust AI output without checking the source?'                   : trust,
    'How well do you understand the limitations of AI tools?'                        : literacy,
    'Has your university given any guidance on responsible AI use?'                  : guidance
})

print(f'Dataset shape  : {df.shape}')
print(f'Missing values : {df.isnull().sum().sum()}')
print(f'Duplicates     : {df.duplicated().sum()}')
print()
df.head()
```

---
## Step 7 — Quick Visualisations

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 9))

df['What region do you attend university'].value_counts().plot(
    kind='bar', ax=axes[0,0], color='#2E75B6', edgecolor='white', rot=20)
axes[0,0].set_title('Responses by Region')
axes[0,0].set_ylabel('Count')

df['How many hours per day do you spend using AI tools?'].value_counts().plot(
    kind='bar', ax=axes[0,1], color='#1E6B3C', edgecolor='white', rot=20)
axes[0,1].set_title('Daily AI Usage Hours')
axes[0,1].set_ylabel('Count')

df['How often do you verify what an AI tool tells you before using it?'].value_counts().sort_index().plot(
    kind='bar', ax=axes[1,0], color='#E15759', edgecolor='white', rot=0)
axes[1,0].set_title('Verify AI Output (1=Never, 5=Always)')
axes[1,0].set_ylabel('Count')

df['Has your university given any guidance on responsible AI use?'].value_counts().plot(
    kind='pie', ax=axes[1,1], autopct='%1.1f%%',
    colors=['#1E6B3C','#E15759','#F0A500'], startangle=90)
axes[1,1].set_title('University AI Guidance')
axes[1,1].set_ylabel('')

plt.suptitle('Survey Response Distributions (N=657)', fontsize=13, y=1.01)
plt.tight_layout()
plt.show()
```

---
## Step 8 — Save to CSV

```python
df.to_csv('data/survey_responses_synthetic.csv', index=False)

print('Saved: data/survey_responses_synthetic.csv')
print(f'Total responses : {len(df)}')
print(f'Columns         : {len(df.columns)}')
print()
print('This file is now ready to be loaded by the main project notebook.')
print('Run CEC420_Project_Neh_Lizza.ipynb next.')
```
